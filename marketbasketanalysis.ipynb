```python
transactions =  data.groupby(['Member_number','itemDescription'])['Quantity'].sum().unstack().reset_index().set_index('Member_number')
```

```python
transactions = transactions.fillna(0)
transactions
```

```python
def encode(x):
    if x <=0:
        return 0
    elif x>=0:
        return 1
basket = transactions.applymap(encode)
```

```python

data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')

data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
data['Weekday'] = data['Date'].dt.day_name()


data.head()
```

```python
# No. of items sold in 2014 and 2015

items_sold_by_year = data[data['Year'].isin([2014, 2015])].groupby('Year').size()

items_sold_by_year
```

```python
# Cummulative day transactions in 2014 & 2015
daily_transactions = (
    data[data['Year'].isin([2014, 2015])]
    .groupby(['Year', 'Date'])
    .size()
    .reset_index(name='Transactions')
)


daily_transactions['Cumulative_Transactions'] = daily_transactions.groupby('Year')['Transactions'].cumsum()

daily_transactions.head()
```

```python
data['Month'] = data['Date'].dt.month

# Monthly quantity
monthly_quantity = data.groupby(['Year', 'Month'])['Quantity'].sum().reset_index()
print(monthly_quantity)
```

```python
data['Weekday'] = data['Date'].dt.day_name()

# Quantity purchased by weekday
weekday_quantity = data.groupby('Weekday')['Quantity'].sum().reset_index()
weekday_quantity = weekday_quantity.sort_values(by='Quantity', ascending=False)
print(weekday_quantity)
```

```python
# Total quantity sold per product
product_quantity = data.groupby('itemDescription')['Quantity'].sum().reset_index()

# Top 10 fast-moving products
top_10_products = product_quantity.sort_values(by='Quantity', ascending=False).head(10)

# Bottom 10 products
bottom_10_products = product_quantity.sort_values(by='Quantity').head(10)

print("Top 10 Fast-Moving Products:")
print(top_10_products)
print("Bottom 10 Products:")
print(bottom_10_products)
```

```python
# Filter data for 2014 and 2015
top_customers = (
    data[data['Year'].isin([2014, 2015])]
    .groupby(['Year', 'Member_number'])
    .size()
    .reset_index(name='Total_Transactions')
)

# Top 10 customers for each year
top_customers_2014 = top_customers[top_customers['Year'] == 2014].nlargest(10, 'Total_Transactions')
top_customers_2015 = top_customers[top_customers['Year'] == 2015].nlargest(10, 'Total_Transactions')

print("Top Customers in 2014:")
print(top_customers_2014)

print("\nTop Customers in 2015:")
print(top_customers_2015)

```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Generate word cloud
product_counts = data['itemDescription'].value_counts()
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(product_counts)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Best Products Wordcloud")
plt.show()
```

```python
import plotly.express as px
import plotly.io as pio

# Prepare data for sunburst
sunburst_data = data.groupby(['Year', 'Month', 'itemDescription'])['Quantity'].sum().reset_index()

# Create sunburst chart
fig = px.sunburst(
    sunburst_data,
    path=['Year', 'Month', 'itemDescription'],
    values='Quantity',
    title="Year-Month-Items Sunburst",
    width=800,
    height=800
)
fig.show()
```
